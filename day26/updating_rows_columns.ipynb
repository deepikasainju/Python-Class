```python
people= {
    "first": ["jon", "jane", "henry", "ken"],
    "last": ["abc","xyz","pqr","def"],
    "email":["jon@gmail.com", "jane@gamil.com", "henry@gmail.com", "ken@gmail.com"]
}
```

```python
import pandas as pd
df=pd.DataFrame(people)
df
```

```python
df.columns=["first_name", "last_name", "Email"]
# sabai columns ko name change gareko
df
```

```python
df.columns=[x.upper() for x in df.columns]
# upper case ma change gareko
df
```

```python
df.columns= df.columns.str.replace('_'," ")
#underscore lai space ma change gareko
df
```

```python
df.columns=[x.lower() for x in df.columns]
# lower case ma change gareko
df
```

```python
df.columns= df.columns.str.replace(' ',"_")
# space lai underscore ma change gareko
df
```

```python
df.rename(columns={"email": "Email"})
# column name ailey change vako xaina
df
```

```python
df.rename(columns={"email": "Email"}, inplace=True)
# inplace=true garda matra change hunxa
df
```

```python
df.loc[2]=["henry", "potter", "hp@gmail.com"]
# row ma 2 index ko value change gareko
# row ma change garna lai iloc ra loc nai use garne
df
```

```python
df.loc[2,["first_name", "last_name"]]=["\Harry", "Potter"]
# specific row ra specific column ko value matra change gareko
df
```

```python
df.loc[2,"Email"]="harrypotter@gmail.com"
df
```

```python
# filter use garera update garne
# filter vaneko specific value choose garera tesma changes garne
filt=(df["Email"]=="harrypotter@gmail.com")
df.loc[filt, "last_name"]="ptter"
df
```

```python
def to_upper(email):
    return email.upper()
df["Email"]=df["Email"].apply(to_upper)
df
```

```python
df["Email"]=df["Email"].apply(lambda x:x.lower())
# lower ma lageko lambda use garera
# apply garda whole column ma chnage aauxa
df
```

```python
df["first_name"]=df["first_name"].replace({"jon":"john", "jane":"jenny"})
# reolace garda specific data ma matra change garxa
df
```

```python
# adding / removing rows and columns from a data frame
df["first_name"]+" "+df["last_name"]
# first name ra last name lai jodera naya series dinxa
```

```python
df["full_name"]=df["first_name"]+" "+df["last_name"]
# naya column thapeko
df
```

```python
df.drop(columns=["full_name"], inplace=True)
# colulmn hatako
df
```

```python
df=df.append({"first_name": "ken", "last_name": "def", "Email":"ken@gmail.com"},ignore_index = True)
df
```

```python
new_row={"first_name": "bruce", "last_name": "ghr", "Email":"hah@gmail.com"}
df.loc[len(df)] = new_row
df
```

```python
df.drop(index=3, inplace=True)
df
```

```python
df.drop(index=df[df['Email']=="jane@gmail.com"].index, inplace=True)
# specified matra delete gareko
df
```

```python
df.sort_values(by="last_name", inplace=True)
df
```

```python
df.sort_values(by="last_name", ascending=False, inplace=True)
df
```

```python
df.loc[0,"last_name"]="ghr"
df
```

```python
df.sort_values(by=["last_name","first_name"], ascending=False, inplace=True)
# first ma last name ko anusar sort garxa ani last name same xa vane first name anusar sort garxa
df
```
